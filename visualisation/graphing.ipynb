```python
import pandas as pd

df = pd.read_pickle('../random_agent_testt.pkl')
print(df) 
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = df['ln']
x = x.to_numpy()
plt.plot(x)
```

```python

ret = np.load('../returns_test.npy', allow_pickle=True)
avgrets = []
avglengths = []
maxtile = []
for r in ret:
    avgrets.append(r[0].numpy()[0])
    avglengths.append(r[1])
    maxtile.append(r[2])
```

```python
from matplotlib.ticker import MaxNLocator

fig, ax1 = plt.subplots()
x = np.arange(1, len(avglengths) + 1)
ax2 = ax1.twinx()
ax1.plot(x, avglengths, 'g-')
ax2.plot(x, avgrets, 'b-')
ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

ax1.set_xlabel("# of episodes trained")
ax1.set_ylabel('Average episode length', color='g')
ax2.set_ylabel('Average episode return', color='b')

ax1.scatter(x, [135, 135, 135, 135, 135], color='orange', label='Maximum tile value')
ax1.legend(loc='upper left')
ax1.set_ylim([125, 275])
ax2.set_ylim([1250, 3000])

plot_margin = 0.3

x0, x1, y0, y1 = plt.axis()
plt.axis((x0 - plot_margin,
          x1,
          y0,
          y1))

for i, txt in enumerate(maxtile):
    ax1.annotate(int(txt), (x[i], 135), xytext=(-25,-3.5), textcoords='offset points')

ax1.set_xticklabels(['0','0','25','50','75','100'])
```
